```python
import pandas as pd
import missingno as msno
```

```python
dados_pro = pd.read_csv('produtividade_soja.csv')
```

```python
dados_pro.dtypes # o código está como int
```

```python
dados_pro.head()
```

```python
dados_pro.describe()
```

```python
msno.matrix(dados_pro,figsize=(10,8))
```

```python
dados_pro_01 = dados_pro.drop(['nivel','codigo_ibge','name'], axis=1)
```

```python
dados_pro_01['Target_Produtividade']= dados_pro_01.mean(axis=1,) # fazer a média por linhas 
dados_pro_01
```

```python
dados_pro_01
```

```python
dados_pro_01.drop(['2004','2005','2006'], axis=1)
```

```python
dados_pro_03 = pd.DataFrame(dados_pro_01['Target_Produtividade']).round(2)
```

```python
dados_pro_03
```

```python
dados_pro_target = pd.merge(dados_pro, dados_pro_03[['Target_Produtividade']], left_index=True, right_index=True, how='inner')
dados_pro_target.shape
```

```python
dados_pro_target
```

```python
dados_pro_target_model = dados_pro_target[['codigo_ibge','Target_Produtividade']]
```

```python
dados_pro_target_model.to_csv('dados_pro_target_model.csv')
```
